```python
import numpy as np
import matplotlib.pyplot as plt
```

# Метод наименьших квадратов

Для систем линейных уравнений

### Реализация МНК

```python
class MLS:

    def __init__(self):
        self.theta = None

    def fit(self, x, y):
        n, *d = x.shape
        l = np.c_[np.ones(n), x]

        self.theta = np.matmul(np.matmul(np.linalg.inv((np.matmul(np.transpose(l), l))), np.transpose(l)), y)

    def predict(self, x):
        n, *b = x.shape
        return np.matmul(self.theta, np.transpose(np.c_[np.ones(n), x]))


def calc_mse(y1, y2):
    return np.mean(np.square(y1 - y2))
```

### Демонстрация работы

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = np.expand_dims(data['data'][:, 0], axis=1)
y = data['target']

mls = MLS()
mls.fit(x, y)
print(mls.theta)
print('Ошибка =', calc_mse(y, mls.predict(x)))


plt.scatter(x, y)

_x = np.linspace(np.min(x), np.max(x))
_y = mls.predict(_x)
plt.plot(_x, _y)
```

```python
### Многомерный случай
data = load_iris()
x = data['data']
y = data['target']

mls = MLS()
mls.fit(x, y)
print(mls.theta)
print('Ошибка =', calc_mse(y, mls.predict(x)))
```
